```python
import pandas as pd

train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df.dropna(inplace=True)
test_df.dropna(inplace=True)
```

```python
import numpy as np
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

non_numerical_columns = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']

for i in non_numerical_columns:
    label_encoder.fit(train_df[i])
    train_df[i] = label_encoder.transform(train_df[i])

    label_encoder.fit(test_df[i])
    test_df[i] = label_encoder.transform(test_df[i])


train_df
```

```python
test_df
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
```

```python
y_train = train_df['satisfaction'].values
y_test = test_df['satisfaction'].values
```

```python
X_train = train_df.iloc[:, 2:-1].values
X_test = test_df.iloc[:, 2:-1].values
```

```python
standard = preprocessing.StandardScaler()
standard.fit(X_train)
X_train_standard = standard.transform(X_train)
X_test_standard = standard.transform(X_test)
```

```python
perceptron_model = Perceptron()
perceptron_model.fit(X_train_standard, y_train)
```

```python
y_pred = perceptron_model.predict(X_test_standard)
y_pred
```

```python
y_test
```

```python
accuracy_score(y_test, y_pred)
```

```python
def RBF(X1, X2, d=3):
    return (1+np.dot(X1.T, X2)) ** d
    
```

```python
import time as t
```

```python
samples, features = X_train_standard.shape
alpha = np.zeros(samples)

start = t.time()
for iteration in range(1):
    for i in range(samples):
        if i%100 == 0:
            print(t.time() - start)
        k = np.zeros(samples)
        for j in range(samples):
            k[j] = RBF(X_train_standard[i], X_train_standard[j])

        if np.sign(np.sum(alpha * y_train * k)) != y_train[i]:
            alpha[i] += 1
```
